```python
df_cat = df['ocean_proximity']
df_cat.head()
```

```python
df2 = pd.get_dummies(data=df, columns=["ocean_proximity"]) # 범주형 변수 원핫인코딩
df2
```

```python
%matplotlib inline
import missingno as msno
import matplotlib.pyplot as plt

msno.matrix(df)
msno.bar(df)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median") # 비정규분포이기 때문에 결측값은 중위수로 채우는 게 좋음
imputer.fit(df2)
X = imputer.transform(df2)
df_tr = pd.DataFrame(X, columns=df2.columns)
df_tr.tail()
```

```python
housing_labels = df_tr["median_house_value"].copy() # 종속변수 카피
df_tr.drop("median_house_value", axis=1, inplace=True) # 종속변수 원본 삭제
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df_tr)
scaled_df = scaler.transform(df_tr)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(scaled_df, housing_labels, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

# mean_squared_error 평균제곱오차
# 제곱값이기 때문에 값을 제곱근으로 줄이려고 sqrt() 씀

y_pred = lin_reg.predict(X_test)
rms = np.sqrt(mean_squared_error(y_test, y_pred))
print(rms)
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(X_train, y_train)

y_pred = tree_reg.predict(X_test)
rms = np.sqrt(mean_squared_error(y_test, y_pred))
print(rms)
```

```python
from sklearn.svm import SVR

svm_reg = SVR(kernel="linear")
svm_reg.fit(X_train, y_train)

y_pred = svm_reg.predict(X_test)
rms = np.sqrt(mean_squared_error(y_test, y_pred))
print(rms)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(n_jobs=-1)
forest_reg.fit(X_train, y_train)

y_pred = forest_reg.predict(X_test)
rms = np.sqrt(mean_squared_error(y_test, y_pred))
print(rms)
```

```python
from sklearn.model_selection import GridSearchCV

# 가장 회귀분석 점수가 잘 나온 RandomForestRegressor의 파라미터를 자세히 정해보자

param_grid = [
    {'n_estimators': [3,10,30,50,100], 'max_features': [2,4,6,8]}
]
forest_reg = RandomForestRegressor(random_state=42, n_jobs=-1)
grid_search = GridSearchCV(forest_reg, param_grid, cv=5)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
import matplotlib.pyplot as plt

feature_importances = grid_search.best_estimator_.feature_importances_
plt.bar(range(len(feature_importances)), feature_importances)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# GridSearchCV는 파라미터 값을 자세히 정해야 하는데 RandomizedSearchCV는 정해진 범위 안에서 값을 랜덤으로 뽑아서 다양하게 실험할 수 있음

param_distribs = {
    'n_estimators': randint(low=1, high=100),
    'max_features': randint(low=1, high=8)
}

forest_reg = RandomForestRegressor(random_state=42, n_jobs=-1)
rnd_search = RandomizedSearchCV(forest_reg, param_distributions=param_distribs, cv=5, random_state=42)
rnd_search.fit(X_train, y_train)
```

```python
cvres = rnd_search.cv_results_

for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(mean_score),params)
```

```python
rnd_search.best_estimator_
```

```python
final_model = rnd_search.best_estimator_
y_pred = final_model.predict(X_test)
rms = np.sqrt(mean_squared_error(y_test, y_pred))
print(rms)
```
