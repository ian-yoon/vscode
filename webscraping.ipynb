```python
html = """
<html><body>
<h1>Hello Python</h1>
<p>웹 페이지 분석</p>
<p>웹 스크레이핑</p>
</body></html>
"""

bs = BeautifulSoup(html, 'html.parser')
h1 = bs.html.body.h1
p1 = bs.html.body.p
p2 = p1.next_sibling.next_sibling # next_sibling은 같은 레벨의 태그 중에 다음 요소. 두 번 한 이유는 엔터키=공백이기 때문에 공백 다음의 태그 속 텍스트를 찾기 위해서

print("h1 = " + h1.text) # 그냥 h1 프린트하면 <h1></h1>태그가 같이 달려서 나오기 때문에 h1.text로 출력
print("p1 = " + p1.text)
print("p2 = " + p2.text)

items = bs.find_all('p')
print(items)
```

```python
html = """
<html><body>
<h1 id='title'>Hello Python</h1>
<p id='body'>웹 페이지 분석</p>
<p>웹 스크레이핑</p>
<span>데이터 수집</span>
</body></html>
"""
# id는 문서 내에서 유일한 값으로 찾아내기 쉬움
# class는 여러 태그에 여러 번 반복해서 쓸 수 있음. 그래도 찾기 쉬운 편.

bs = BeautifulSoup(html, 'html.parser')
title = bs.find(id="title")
body = bs.find(id="body")
span = bs.find("span")

print("#title =", title.text)
print("#body =", body.text)
print("#span =", span.text)
```

```python
html = """
<html><body>
<ul>
<li><a href="http://naver.com">naver</a></li>
<li><a href="http://daum.net">daum</a></li>
<li><a href="http://google.com">google</a></li>
<li><a href="http://nate.com">nate</a></li>
<li><a href="http://yahoo.com">yahoo</a></li>
</ul>
</body></html>
"""

bs = BeautifulSoup(html, 'html.parser')
links = bs.find_all("a")
for a in links:
    href = a['href']
    text = a.text
    print(text, ":", href)

# 태그의 속성을 출력하려면 태그["속성"] ex) href=a['href']
```

```python
from bs4 import BeautifulSoup
import urllib.request as req

url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
res = req.urlopen(url)
bs = BeautifulSoup(res, "html.parser")
title = bs.find("title").text
wf = bs.find("wf").text
print(title)
print(wf)

result1 = []
result2 = []
result3 = []
result4 = []
result5 = []
items = bs.find_all('data')
for item in items:
    times = item.find_all('tmef')
    wfs = item.find_all('wf')
    tmns = item.find_all('tmn')
    tmxs = item.find_all('tmx')
    rnsts = item.find_all('rnst')
    for idx in range(len(times)):
        result1.append(times[idx].text)
        result2.append(wfs[idx].text)
        result3.append(tmns[idx].text)
        result4.append(tmxs[idx].text)
        result5.append(rnsts[idx].text)

result = zip(result1, result2, result3, result4, result5)
rows = list(result)
rows
```

```python
import pandas as pd
df = pd.DataFrame(rows, columns=['일시', '날짜', '최저', '최고', '강수확률'])
df.head()
```

```python
df.to_csv('c:/data/weather/data.csv', encoding='ms949', index=False)
```

```python
from bs4 import BeautifulSoup
import urllib.request as req

url = "http://finance.naver.com/"
res = req.urlopen(url)
bs = BeautifulSoup(res, "html.parser")

result1 = []
moneys = bs.select('div.section1 > div.group1 > table > tbody > tr > th > a')
for money in moneys:
    result1.append(money.text.strip())

result2 = []
exchange = bs.select('div.section1 > div.group1 > table > tbody > tr > td:nth-child(2)')
for ex in exchange:
    result2.append(ex.text)

result = list(zip(result1, result2))
result
```

```python
from bs4 import BeautifulSoup
import urllib.request as req
import urllib.request
from urllib.request import urlopen

codes = []
for num in range(51,53):
    url = "http://www.auction.co.kr/category/category{0}.html".format(num)
res = req.urlopen(url)
bs = BeautifulSoup(res, "html.parser", from_encoding='ms949')

category = bs.select("#navCat > div > ul > li > a")
for cate in category:
    name = cate.text
    href = cate.attrs["href"].split("=")[1] # http 주소의 제일 마지막에 list?category=상품코드 중 상품코드를 추출하기 위해 =를 기준으로 해서 나눔
    codes.append(cate)
    print("상품분류명: " + name)
    print("분류코드: " + href + "\n")
```
