```python
import nltk
nltk.download("book", quiet=True)
from nltk.book import *

nltk.corpus.gutenberg.fileids()
```

```python
emma_raw = nltk.corpus.gutenberg.raw("austen-emma.txt")
print(emma_raw[:500])
```

```python
from nltk.tokenize import sent_tokenize

# 3번 문장
print(sent_tokenize(emma_raw[:1000])[3])
```

```python
from nltk.tokenize import word_tokenize

# 50~99 단어
word_tokenize(emma_raw[50:100])
```

```python
from nltk.stem import PorterStemmer, LancasterStemmer

st1 = PorterStemmer()
st2 = LancasterStemmer()

words = ["fly", "flies", "flying", "flew", "flown"]

print([st1.stem(w) for w in words])
print([st2.stem(w) for w in words])
```

```python
# 원형 복원
from nltk.stem import WordNetLemmatizer
nltk.download('omw-1.4')

lm = WordNetLemmatizer()
[lm.lemmatize(w, pos="v") for w in words]
```

```python
%matplotlib inline
from nltk import Text
import matplotlib.pyplot as plt
from nltk.tokenize import RegexpTokenizer

retokenize = RegexpTokenizer("[\w]+")

text = Text(retokenize.tokenize(emma_raw))
plt.rcParams["figure.figsize"]=(15,10)
plt.rcParams["font.size"]=15

text.plot(20)

plt.show()
```

```python
text.dispersion_plot(["Emma", "Knightley", "Frank", "Jane", "Harriet", "Robert"])
```

```python
text.concordance("Emma")
```

```python
text.similar("Emma")
```

```python
text.common_contexts(["Emma", "she"])
```

```python
from nltk import FreqDist
from nltk.tag import pos_tag

stopwords = ["Mr.", "Mrs.", "Miss", "Mr", "Mrs", "Dear"]
emma_tokens = pos_tag(retokenize.tokenize(emma_raw))
names_list = [t[0] for t in emma_tokens if t[1]=="NNP" and t[0] not in stopwords]
fd_names = FreqDist(names_list)
```

```python
fd_names.N(), fd_names["Emma"], fd_names.freq("Emma")
```

```python
fd_names.most_common(5)
```

```python
from wordcloud import WordCloud

wc = WordCloud(width=1000, height=600, background_color="white", random_state=0)
plt.imshow(wc.generate_from_frequencies(fd_names))
plt.show()
```

```python
f = open("/Users/ian/Desktop/Study/data/text/trumph.txt", "r", encoding="cp949")
lines = f.readlines()
f.close()
```

```python
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords

tokenizer = RegexpTokenizer("[\w]+")
stop_words = stopwords.words("english")
words = lines[0].lower()
tokens = tokenizer.tokenize(words)
stopped_tokens = [i for i in list(tokens) if not i in stop_words]
stopped_tokens2 = [i for i in stopped_tokens if len(i)>1]
```

```python
import pandas as pd

pd.Series(stopped_tokens2).value_counts().head(10)
```

```python
%matplotlib inline
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt

wordcloud = WordCloud(
    width = 800,
    height = 800,
    background_color = "white"
)

count = Counter(stopped_tokens2)
print(count)
wordcloud = wordcloud.generate_from_frequencies(count)

fig = plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
plt.show()
```
