```python
from sklearn import datasets
iris = datasets.load_iris()
print(iris['feature_names'])
```

```python
import numpy as np
import pandas as pd
df = pd.DataFrame(data=np.c_[iris.data, iris.target], columns=iris['feature_names']+['target']) # np.c_는 결합 함수
df['target']=df['target'].map({0:'setosa', 1:'versicolor', 2:'virginica'})
df.head()
```

```python
df["target"].value_counts() # 품종별 샘플 수
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

colors = ["red", "green", "blue"]

sns.countplot('target', data=df, palette=colors)
plt.title('분류\n (0:setosa, 1:versicolor, 2:virginica)', fontsize=12)
```

```python
import pandas as pd
df = pd.read_csv("c:/data/traffic/data.csv", encoding='ms949')
df.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plot
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

plt.style.use('ggplot')
months = ['1월', '2월', '3월', '4월', '5월', '6월', '7월', '8월', '9월', '10월', '11월', '12월']
months_index = range(1, len(months)+1)
result = []
for i in range(12):
    result.append(df[str(i+1)+'월'].sum() / 10000000)
```

```python
result
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(1,1,1)
ax1.bar(months_index, result, align='center', color='darkblue')
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
plt.xticks(months_index, months, rotation=0)
plt.tick_params(labelsize=12)
plt.xlabel('월', fontsize=12)
plt.ylabel('합계', fontsize=12)
plt.title('통계', fontsize=14)
plt.savefig('c:/data/images/bar_plot.png', dpi=400, bbox_inches='tight')
plt.show()
```
