```python
# 유사도 (거리) 계산
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

X = [10,25,47,68,89]
y = [20,40,55,65,75]

plt.figure(figsize=(5,5))
plt.scatter(X,y)
mypoint=(80,5)
plt.scatter(mypoint[0],mypoint[1],s=100,c='r')
```

```python
# 유클리드 거리 구하기

from math import sqrt

def distance(p, m):
    items=[]
    for i in range(len(p-1)):
        distance = sqrt((m[0] - p[i][0])**2 + (m[1] - p[i][1])**2)
        items.append(distance)
    return items

points = np.array(list(zip(X,y)))
r = distance(points, mypoint)
print(r)
m = min(r)
i = np.argmin(r)
print('가장 가까운 포인트의 인덱스 : ',i)
print('가장 가까운 포인트 : ',points[i])
```

```python
# 유클리드 거리 구하기 (함수 이용)

from scipy.spatial import distance

results = []
for point in points:
    dist = distance.euclidean(point, mypoint)
    results.append(dist)

results
```

```python
# 맨해튼 거리 구하기

def distance(p, m):
    items = []
    for i in range(len(p-1)):
        distance = abs(m[0] - p[i][0]) + abs(m[1] - p[i][1])
        items.append(distance)
    return items

r = distance(points, mypoint)
print(r)
m = min(r)
i = np.argmin(r)
print('가장 가까운 포인트의 인덱스 : ',i)
print('가장 가까운 포인트 : ',points[i])
```

```python
# 맨해튼 거리 구하기 (함수 이용)

from scipy.spatial import distance

results = []
for point in points:
    dist = distance.cityblock(point, mypoint)
    results.append(dist)
results
```
