```python
def my_zip(a,b):
    if len(a) != len(b):
        print('입력값의 길이가 맞지 않습니다.')
    else:
        n = len(a)
        c = []
        for i in range(n):
            c.append((a[i],b[i]))
        return c

my_zip(a,b)
```

```python
# 은행 예금 이율은 3%, 금액=원금*이율**연수

def saving(money, years, rate):
    result = money*(1 + rate / 100)**years
    return result

money = 10000
years = np.arange(0,11)
rate = [3, 5, 7]
total_money = np.zeros((len(rate), len(years))) # 3행 10열의 2차원 배열에 0 채우기

for i in range(len(rate)): # 이자율
    for j in range(len(years)): # 연수
        total_money[i,j] = saving(money, years[j], rate[i]) # 저축액 계산
    plt.plot(years, total_money[i,:], label='{}%'.format(rate[i]))

plt.legend(fontsize = 12)
plt.grid()
plt.xlabel('Years', fontsize = 12)
plt.ylabel('Total money', fontsize = 12)
```

```python
x = [3, 1, 5, 4, 3, 9, 8, 8, 7, 8, 3, 4, 6, 6, 8, 8, 1, 2, 2]

# 평균
print(np.average(x))

# 중앙값(중위수)
x.sort()
print(np.median(x))

# 최빈값
bins = np.bincount(x)
idx = bins.argmax()
print(idx)

print(np.bincount(x))
```

```python
# 편차 = 변량 - 평균
# 분산 = 편차를 제곱한 값들의 평균
# 표준편차 = 분산의 제곱근

x = 0, 1, 3, 6, 12, 13, 10, 7, 5, 1

mean = np.average(x)
print(mean)
print(np.var(x)) # 분산
print(np.std(x)) # 표준편차
```

```python
variance = 0
for i in range(len(x)):
    variance += (x[i]-mean)**2

variance /= len(x)
std = np.sqrt(variance)

print('분산: ', variance)
print('표준편차: ', std)
```
